```python
import pickle
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
with open('merge_df.pickle', 'rb') as f:
    merge_df = pickle.load(f)
```

```python
def lsttostrlst(lst_lst):
    merge_str=[]
    for lst in lst_lst:
        merge_str.append(' '.join(lst))
    return merge_str
```

```python
count_vectorizer_stem = CountVectorizer(ngram_range=(1, 1), analyzer='word', max_df=0.9, max_features = 1000,
                                        stop_words='english', binary=False, tokenizer=str.split)
```

```python
str_stemmed = lsttostrlst(merge_df['tweet_stemmed'])
bag_of_words_stem = count_vectorizer_stem.fit_transform(str_stemmed)

feature_names = count_vectorizer_stem.get_feature_names()
pd.DataFrame(bag_of_words_stem.toarray(), columns = feature_names)
```

```python
with open('corpus', 'r') as f:
    corpus = f.read()
```

```python
labels, texts = [], []
for i, line in enumerate(corpus.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))
```

```python
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model
from sklearn.metrics import accuracy_score

train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'], test_size=0.30)

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)

count_vect = CountVectorizer(ngram_range=(1, 1), analyzer='word', max_df=0.9, max_features = 1000, 
                             stop_words='english', binary=False)

count_vect.fit(trainDF['text'])

xtrain_count =  count_vect.transform(train_x)
xvalid_count =  count_vect.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions = classifier.predict(xvalid_count)
predictions
```

```python
accuracy_score(valid_y, predictions)
```

```python
count_vect = TfidfVectorizer(ngram_range=(1, 2), analyzer='word', max_df=0.9, max_features = 6000,
                             stop_words='english', binary=False)

count_vect.fit(trainDF['text'])

xtrain_count =  count_vect.transform(train_x)
xvalid_count =  count_vect.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions = classifier.predict(xvalid_count)
predictions
```

```python
accuracy_score(valid_y, predictions)
```
